## 11. Identify and remove outliers in the km_driven column using the IQR method?

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]
print(f"Outliers:\n{outliers}")

# Remove outliers
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]
print(f"\nData without outliers:\n{df_no_outliers}")
```

### 6. What are the most popular product keywords?

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import string
import nltk
nltk.download('stopwords')

# Define a function to clean and extract keywords from text
def clean_and_tokenize(text):
    # Convert to lowercase, remove punctuation and stopwords
    stop_words = set(stopwords.words('english'))
    text = text.lower()
    text = ''.join([char if char not in string.punctuation else ' ' for char in text])  # remove punctuation
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return words

# Extract keywords from both product_name and about_product
df['keywords'] = df['product_name'].apply(clean_and_tokenize) + df['about_product'].apply(clean_and_tokenize)

# Flatten the list of keywords and count their frequency
all_keywords = [keyword for sublist in df['keywords'] for keyword in sublist]

# Create a DataFrame with word counts
word_counts = pd.Series(all_keywords).value_counts()

# Get the most popular 10 keywords
top_keywords = word_counts.head(10)

print("Top 10 Most Popular Product Keywords:")
print(top_keywords)
```
